```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022 (1).xlsx')
#df
```

```python
#Verificamos información del DataFrame
df.info()
```

**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

Filtro 1: por objeto

La idea tras la selección de esta instancia através de esta columna de "Nombre Vendedor" es aislar las ventas que esta persona tiene así como el resto de los datos de cada venta que tuvo. Nos permite ver con detalle la lista de todos los movimientos que esta persona generó y este dataset puede servinirnos para comparar el desempeño que esta vendedora tiene.

```python
filtro1=df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

Filtro 2: por filas

En ocasiones no siempre quieres ver toda la información. En este caso podemos usar este filtro para solo exportar las filas de ciertas fechas.

```python
filtro2= df.iloc[10:20,: ]           #Filas consecutivas
filtro2
```

Filtro 3: por columnas

Así como con las filas, puede que necesites hacer un análisis de solo algunas columnas y con este filtro puedes tener un archivo más limpio.

Aquí por ejemplo, me interesaba aislar el costo y margen de los productos y su relación entre ellos. Esto puede ser usado más adelante para encontrar algún tipo de cluster entre items.

```python
filtro3=df.iloc[ : , [7,8,9,13]]
filtro3
```

Filtro 4: filtro de filas vs columnas

La conjunción de los anteriores. Aquí lo que quiero ver es los valores que me arroja al aislar el margen del público en general. Con esto puedo ver que tanta varianza tiene y poder empezar a identificar outliers.

```python
# Nuevo DF con el Nombre de los clientes como la columna índice 
df1= pd.read_excel('Detalle precios y productos fabricados 2022 (1).xlsx', index_col=3)
```

```python
filtro4=df1.loc[["PÚBLICO EN GENERAL"] , ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

Filtro 5: por cabecera

El filtro tiene varios usos. Principalmente, creo que nos permite visualizar el tipo de datos que tiene cada columna y para después empezar a tratarlos.

```python
filtro5=df.head(3)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

Filtro 6: por comparación

Aquí establezco un monto para poder aislar a los que solo cumplen con la condición del margen, que en este caso sean de 300 en adelante. Esto me permite ver por ejemplo, que vendedores traen una mayor cantidad de ventas a partir del monto que elegí.

```python
#Filtro por comparacion
filtro6=df[df["MARGEN_TOTAL_CALCULADO"] > 300]
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

Filtro 7= "Y"

La selección de estas columnas me permite aislar solamente las compras superiores al margen pero solamente las que superan el monto que establecí. Así como en el caso de antes, solo que ahora no desperdicio mi tiempo viendo márgenes más bajos.

```python
filtro7=df[(df["MARGEN_TOTAL_CALCULADO"] > 300) & (df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro7
```

Filtro 8: "O"

Este une el filtro 6 y 7 incluyendo todos los datos que me interesan.

```python
filtro8=df[(df["MARGEN_TOTAL_CALCULADO"] > 300) | (df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro8
```

Filtro 9: "~" (not)

Y esto me permite ver lo contrario a la selección anterior.

```python
#Filtro not ~
filtro9=df[~ (df["MARGEN_TOTAL_CALCULADO"] > 300) & ~ (df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro9
```
